## Step 2: Dataset Structure and Creation

### Expected Dataset Structure:
- **experience**: Years of professional experience (numeric)
- **test_score**: Performance test score out of 100 (numeric)
- **department**: Employee department (categorical, optional)
- **education_level**: Highest degree obtained (categorical, optional)
- **salary**: Annual salary in USD (target variable)

For this tutorial, we'll create a synthetic dataset. In production, you would load your actual data using `pd.read_csv('your_data.csv')`

```python
# Create synthetic dataset for demonstration
np.random.seed(42)

n_samples = 200

# Generate features
experience = np.random.uniform(0, 15, n_samples)
test_score = np.random.uniform(50, 100, n_samples)
department = np.random.choice(['Engineering', 'Sales', 'Marketing', 'HR'], n_samples)
education_level = np.random.choice(['Bachelor', 'Master', 'PhD'], n_samples)

# Generate salary with realistic relationships
# Salary = base + (experience * weight) + (test_score * weight) + noise
salary = (
    40000 + 
    (experience * 3500) + 
    (test_score * 400) + 
    np.random.normal(0, 5000, n_samples)
)

# Create DataFrame
df = pd.DataFrame({
    'experience': experience,
    'test_score': test_score,
    'department': department,
    'education_level': education_level,
    'salary': salary
})

# Introduce some missing values to simulate real-world data
missing_indices = np.random.choice(df.index, size=10, replace=False)
df.loc[missing_indices, 'test_score'] = np.nan

print(f"Dataset created with {len(df)} samples")
print(f"\nDataset shape: {df.shape}")
print(f"\nFirst 5 rows:")
df.head()
```
